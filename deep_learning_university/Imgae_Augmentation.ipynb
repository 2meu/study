```python
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
X_train = X_train.astype('float32')
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2],
                         1)
```

```python
datagen = ImageDataGenerator(rotation_range=120, rescale=1./255,
                            data_format='channels_last')
```

```python
datagen.fit(X_train)
```

```python
i = 0

for X_batch, Y_batch in datagen.flow(X_train, Y_train, batch_size=32):
    if i == 0:
        AugX = X_batch
        AugY = Y_batch
        i += 1
        print('i={}'.format(i))
    AugX = np.append(AugX, X_batch, axis=0)
    AugY = np.append(AugY, Y_batch, axis=0)
    i+=1
    print("i={}".format(i))
    if i >= len(X_train)/32:
        break
```

```python
for i in range(0,9):
    plt.subplot(330 + 1 + i)
    plt.imshow(AugX[i].reshape(28,28),cmap=plt.get_cmap('gray'))
    
plt.show()
```
